Welcome to the Simple Number Evaluator!
Input numbers, and then it determines whether each number is even or odd. After that, it computes the total and average of the numbers.

```python
def function(num, input):

    if num % 2 == 0:
        print(f"The number", num, "is even.")
    else:
        print(f"The number", num, "is odd.")
    input.append(num)

num_count = int(input("How many numbers will you enter? "))

nums = []

for i in range(num_count):
    number = int(input(f"Enter number {i + 1}: "))
    function(number, nums)

total = sum(nums)

average = total/num_count

print(f"The total of the numbers entered is", total)
print(f"The average of these numbers is", average)

print("Thanks for using the evaluator! =P")
```
